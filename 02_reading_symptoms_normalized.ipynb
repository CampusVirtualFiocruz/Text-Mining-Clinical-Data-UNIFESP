```python
import pandas as pd
import re
```

```python
df = pd.read_excel("leitura_manual_estudantes.xlsx", sheet_name='Contagem')
```

```python
def remove_text_before_key_strings(text):
    # Define the regex pattern to match the key strings in a case-insensitive manner
    pattern = re.compile(r'.*?(atual:|evolução:|hoje)', re.IGNORECASE)
    
    # Find the first occurrence of the key strings and remove everything before it
    match = pattern.search(text)
    if match:
        return text[match.start():]
    else:
        return "vazio"  # Key string not found, return the original text
```

```python
df['hpma_normalizada'] = None 
```

```python
for i in range(len(df['hpma'])):
    texto = str(df.iat[i,1])
    df.iat[i,37] = remove_text_before_key_strings(texto)
```
